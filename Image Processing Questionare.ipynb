## Location of Green Boxes 

```python
import cv2
import numpy as np
import pandas as pd

# Load the image
image = cv2.imread('C:/Users/Hassan/Desktop/Projects/ResponseLetter/key.png')

# Crop 400 pixels from the top of the image
image = image[400:, :]

# Convert the image from BGR to HSV color space
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define the range of green color in HSV
lower_green = np.array([40, 40, 40])
upper_green = np.array([70, 255, 255])

# Create a mask for the green color
mask = cv2.inRange(hsv, lower_green, upper_green)

# Find the contours in the mask
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Create an empty list to store the correct answers
correct_answers = []

# Iterate over each contour
for contour in reversed(contours):
    # Compute the bounding rectangle for the contour
    x, y, w, h = cv2.boundingRect(contour)
    
    # Compute the center of the bounding rectangle
    cx = x + w // 2
    cy = y + h // 2
    
    # Compute the column index for the correct answer
    col = (cx - 50) // 100
    
    # Compute the row index for the correct answer
    row = (cy - 50) // 100
    
    # Store the row and column indices in the correct_answers list
    correct_answers.append({
        'row': row,
        'col': col
    })

# Create a DataFrame from the correct_answers list
df = pd.DataFrame(correct_answers)

# Reset the index of the DataFrame to start from 1
df.index = df.index + 1

# Define a custom function to compute the value of the new column
def find_the_answer(row):
    if 1 <= row['col'] <= 3:
        return 1
    elif 5 <= row['col'] <= 7:
        return 2
    elif 9 <= row['col'] <= 11:
        return 3
    elif 13 <= row['col'] <= 16:
        return 4
    else:
        return row['col']

# Apply the custom function to each row in the DataFrame
df['answer'] = df.apply(find_the_answer, axis=1)

# Save the DataFrame to a CSV file with an index column
df.to_csv('C:/Users/Hassan/Desktop/Projects/ResponseLetter/key.csv', index=True, index_label='question')

# Print a message to indicate that the data has been saved
print('The locations of the correct answers have been saved to the "correct_answers.csv" file with a new column.')
```

## Correct Answers 

```python
import cv2
import numpy as np
import pandas as pd

# Load the image
image = cv2.imread('C:/Users/Hassan/Desktop/Projects/ResponseLetter/key.png')

# Crop 400 pixels from the top of the image
image = image[400:, :]

# Convert the image from BGR to HSV color space
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define the range of green color in HSV
lower_green = np.array([40, 40, 40])
upper_green = np.array([70, 255, 255])

# Create a mask for the green color
mask = cv2.inRange(hsv, lower_green, upper_green)

# Find the contours in the mask
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Create an empty list to store the correct answers
correct_answers = []

# Define the minimum area for a contour to be considered a fully green cell
min_area = 500

# Iterate over each contour
for contour in reversed(contours):
    
    x, y, w, h = cv2.boundingRect(contour)
    cx = x + w // 2
    cy = y + h // 2
    
    col = (cx - 50) // 100
    row = (cy - 50) // 100

    # Check if the area of the contour is above the minimum area
    if cv2.contourArea(contour) >= min_area:

        # Store the row and column indices in the correct_answers list
        correct_answers.append({
            'row': row,
            'col': col
        })
    else: 
        correct_answers.append({
            'row': row,
            'col': 0, 
        }) 
        

# Create a DataFrame from the correct_answers list
df = pd.DataFrame(correct_answers)
df.index = df.index + 1
df['correct'] = df.apply(find_the_answer, axis=1)

# Save the DataFrame to a CSV file
df.to_csv('C:/Users/Hassan/Desktop/Projects/ResponseLetter/correct_answers.csv', index=True, index_label='question')

# Print a message to indicate that the data has been saved
print('The locations of the correct answers have been saved to the "correct_answers.csv" file.')
```

### Detecting Answers in Answer Sheet + Save in CSV

```python
import cv2
import numpy as np
import csv

# Load the image
image = cv2.imread('C:/Users/Hassan/Desktop/Projects/ResponseLetter/image0000013A.tif')
image = image[600:-200, 80:-80]
image = cv2.resize(image, (width, height))

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply adaptive thresholding
thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 4)

# Perform morphological operations
kernel = np.ones((3, 3), np.uint8)
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=1)

# Find contours
contours, _ = cv2.findContours(opening, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Filter contours based on area and aspect ratio
filtered_contours = []
for contour in contours:
    area = cv2.contourArea(contour)
    x, y, w, h = cv2.boundingRect(contour)
    aspect_ratio = w / float(h)

    # Filter based on area and aspect ratio
    if 40 < area < 120 and 0.5 < aspect_ratio < 2:
        filtered_contours.append((x, y, x + w , y + h))

# Draw rectangles around the filtered contours
for (x, y, x2, y2) in filtered_contours:
    cv2.rectangle(image, (x, y), (x2, y2), (0, 255, 0), 2)

    # Crop the regions within the rectangles
    cropped_region = image[y:y2, x:x2]


# Save the locations to a CSV file
csv_file = 'C:/Users/Hassan/Desktop/Projects/ResponseLetter/answers_sheet.csv'
with open(csv_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['x', 'y', 'x2', 'y2'])  # Write the header
    for (x, y, x2, y2) in reversed(filtered_contours):
        writer.writerow([x, y, x2, y2])

# Display the image with rectangles
cv2.imshow('Answer Sheet', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
